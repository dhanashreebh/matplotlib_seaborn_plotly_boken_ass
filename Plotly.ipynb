```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Set a random seed for reproducibility
np.random.seed(30)

# Generate the dataset
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
ax.scatter(df['X'], df['Y'], df['Z'], c='b', marker='o', alpha=0.6)

# Adding titles and labels
ax.set_title('3D Scatter Plot of Data Points')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

# Show grid
ax.grid(True)

# Display the plot
plt.show()
```

volin plot for student grades

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(15)

# Generate student grades data
data_grades = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}
df_grades = pd.DataFrame(data_grades)

# Create a violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='Grade', y='Score', data=df_grades, inner='quartile', palette='muted')

# Adding titles and labels
plt.title('Distribution of Scores Across Different Grade Categories')
plt.xlabel('Grade')
plt.ylabel('Score')

# Show the plot
plt.show()
```

Heat map for sales

```python
# Set random seed for reproducibility
np.random.seed(20)

# Generate sales data
data_sales = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df_sales = pd.DataFrame(data_sales)

# Pivot the sales data to create a matrix for the heatmap
sales_matrix = df_sales.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='sum', fill_value=0)

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(sales_matrix, annot=True, fmt=".0f", cmap='YlGnBu', cbar_kws={'label': 'Total Sales'})

# Adding titles and labels
plt.title('Sales Heatmap Across Months and Days')
plt.xlabel('Month')
plt.ylabel('Day')

# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create a DataFrame (optional)
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create a 3D surface plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')

# Adding titles and labels
ax.set_title('3D Surface Plot of z = sin(sqrt(x^2 + y^2))')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(25)

# Generate the dataset
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Create a bubble chart
plt.figure(figsize=(10, 6))
bubble = plt.scatter(df['GDP'], df['Population'],
                     s=df['Population']*10,  # Bubble size
                     alpha=0.6,
                     edgecolors="w",
                     linewidth=2)

# Adding titles and labels
plt.title('Bubble Chart of Population vs GDP by Country')
plt.xlabel('GDP (in millions)')
plt.ylabel('Population (in thousands)')

# Annotate each bubble with the country name
for i in range(len(df)):
    plt.annotate(df['Country'][i], (df['GDP'][i], df['Population'][i]),
                 textcoords="offset points",
                 xytext=(0,10),
                 ha='center')

# Show the plot
plt.grid()
plt.show()
```
